```python
def average_by_key(rows):
    score = [r["score"]
            for r in rows if "score" in r]
    if not score:
        return 0.0
    total = sum(score)
    return total / len(score)

print(average_by_key([{"score": 5}, {"score": 7}]))  # Expected Output: 6.0
print(average_by_key([]))   # Expected Output: 0
```

```python
def even_numbers(nums):
    evens = []
    for n in nums:
        if n % 2 == 0:
            evens.append(n)
    return evens
print(even_numbers([1,2,3,4,5,6])) # Expected Output: [2, 4, 6]
```

```python
class Artist:
    def __init__(self, name, fee):
        self.name = name
        self.__fee = fee #private attribute, do not change this.

    def add_charges(self, additional_charges):
        self.__fee = self.__fee + additional_charges
    def get_fee(self):
       return self.__fee
a1 = Artist("Taylor Swift", 50)
a2 = Artist("Lisa", 65)
a1.add_charges(20)
a2.add_charges(20)
print("Taylor Swift fee=", a1.get_fee()) # Expected Output: 70
print("Lisa fee=", a2.get_fee()) # Expected Output: 85
```

Develop a python program to create text file with three lines. (File name - Student_ID_CVE.txt)

Hello! Welcome to KMUTT.

We are Civil Engineering Students.

We don't want to learn programming T_T

```python
# Write your program here
file_name = "Student_ID_CVE.txt"
file_content = ["Hello! Welcome to KMUTT.\n",
                "We are Civil Engineering Students.\n",
                "We don't want to learn programming T\n_T"]
with open(file_name, "w", encoding="utf-8") as f:
    f.writelines(file_content)
print(f"File '{file_name}' has been created!!!")
```



Develop a python program to read previous text file line by line and print second line.

Expected output:
We are Civil Engineering Students.





```python
# Write your program here
file_name = "Student_ID_CVE.txt"
try:
    with open(file_name, "r", encoding="utf-8") as f:
        lines = f.readlines()
        if len(lines) >= 2:
            print(lines[1].strip())
        else:
            print("File contains less than two lines.")
except FileNotFoundError:
    print(f"Unexpected error occured: file '{file_name}' cannot be found. Create again before running again.")
```
